```julia
using PyPlot
```

```julia
lyap = [] #Se crea un vector vacío.
list = [] #Se crea un vector vacío.
for m in 3:0.001:5 #Se realiza un for donde se calculan los valores de Lyapunov.
    xn = rand() #Se asigna un valor aleatorio a xn.
    lyp = 0 #Se inicia el valor de lyp.

    for i in 1:10000 #Se obtiene el valor de xn mediante el mapeo logístico.
        
        xn = m*xn*(1 - xn) 
       
        if i>300 #Se obtiene la suma del logaritmo.
            lyp += log(abs(m-2*m*xn)) 
       
        end
    end   
    push!(list,xn) #Se guarda en list el valor de xn.    
    lyp = lyp/10000 #Se divide entre 10000 el valor obtenido del exponente.
    push!(lyap,lyp) #Se guarda en lyap el valor obtenido del exponente.
end

m = 3:0.001:5
```

```julia
plot(m,lyap)  #Se grafica el exponente de Lyapunov.
xlabel("Lyapunov");ylabel("m")
title("Exponente de Lyapunov")
```

```julia
plot(m,list,".") #Se grafican las bifurcaciones.
xlabel("m");ylabel("xn")
title("Bifurcaciones")
```

## Ejercicio 1

```julia
function metodo_euler(f,x0,t0,tf,h) #Se crea una función que permite, mediante el método de Euler, obtener
    #la solución aproximada de una ecuación diferencial. Dicha función toma como entradas la función que 
    #aparece en tal ecuación, la condición inicial, el valor inicial de t, el valor final de t, y el valor de h.
    listt = [] #Se crea un vector cuyo valor inicial es t0 y valor final tf donde la separación entre  
    #sus valores está dada por el valor de h.
        for i in t0:h:tf
            push!(listt,i) 
        end
    listx = [] #Se crea un vector en donde se guardan los valores xk del método de Euler.
    xk = x0 
        for j in 1:length(listt)
            push!(listx,xk)
            xk = xk + f(xk,listt[j])*h        
        end
    return listt, listx #Se regresan la listas listt y listx.
end
```

```julia
function fu(x,t) #Se crea una función particular para implementar el método de Euler.
    return -2.5*x
end
```

```julia
function exponencial(t) #Se crea la función exponencial pues es la solución exacta y será graficada junto con
    #la aproximación del método de Euler para ver gráficamente la comparación entre estas soluciones.
    return exp(-2.5t)
end    
```

```julia
t=[] #Se grafica sólamente la exponencial.
for i in 0:0.001:8 
    push!(t,i) 
end
plot(t, [exponencial(i) for i in t])
```

```julia
listt,listx = metodo_euler(fu,1,0,8,1.0) #Se grafica el método de Euler para h = 1.0, así como la exponencial.
plot(listt,listx)
t=[] 
for i in 0:0.001:8 
    push!(t,i) 
end

plot(t, [exponencial(i) for i in t])
```

```julia
#Se grafica el método de Euler para h = 0.5, 0.2, 0.01 así como la función exponencial.
listt,listx = metodo_euler(fu,1,0,8,0.5) #Para h = 0.5 se tiene la gráfica azul.
plot(listt,listx)
listt,listx = metodo_euler(fu,1,0,8,0.2) #Para h = 0.2 se tiene la gráfica naranja.
plot(listt,listx)
listt,listx = metodo_euler(fu,1,0,8,0.01) #Para h = 0.01 ya no se alcanza a distinguir el color de la gráfica
#pues es bastante parecida a la gráfica de la exponencial, entonces las gráficas se empalman.
plot(listt,listx)
t=[] 
for i in 0:0.001:8 
    push!(t,i) 
end
plot(t, [exponencial(i) for i in t]) #Para la exponencial se tiene la gráfica roja.
#Se observa que para valores de h cada vez más pequeños, el método de Euler es una mejor aproximación a la 
#solución exacta.
```

## Ejercicio 2

El sistema físico modelado por dicha ecuación es el oscilador armónico.

Mediante la ecuación característica, determinamos la solución general de la ecuación.
$$ \lambda^2 = - k $$

$$ x(t) = A e^{\sqrt{-k}t} + B e^{-\sqrt{-k}t} $$

Evaluando en t = 0
$$ x(t = 0) = A + B $$

$$ x'(t = 0) = \sqrt{-k}(A-B) $$

Resolviendo el sistema de ecuaciones para hallar A y B en términos de x(0) y x'(0) se tiene

$$ A = {x(0) \over 2} + {x'(0) \over 2\sqrt{-k}} $$

$$ B = {x(0) \over 2} - {x'(0) \over 2\sqrt{-k}} $$

Por lo que la solución general en términos de las condiciones iniciales x(0) y x'(0) es

$$ x(t) = \displaystyle{\left( {x(0) \over 2} + {x'(0) \over 2\sqrt{-k}} \right)} e^{\sqrt{-k}t} + \displaystyle{\left( {x(0) \over 2} - {x'(0) \over 2\sqrt{-k}} \right)} e^{-\sqrt{-k}t} $$


## Ejercicio 3

Haciendo

$$ u = \dot x(t) $$

Derivando, se tiene

$$ \dot u = \ddot x(t) $$

Pero, como    $ \ddot x = - k x $ 

Entonces, tenemos el siguiente sistema de ecuaciones

$$ \dot u = - k x $$
$$ \dot x = u$$

Por lo que podemos utilizar el método de Euler en su forma vectorial

$$ \frac{d\overline{x}}{dt} = \overline{f}(\overline{x},t)$$

donde $ \overline{x} = (x,u) $ $  $ y $  $  $\overline{f}(\overline{x},t) = (\dot x, \dot u) = (u,-kx) $.


## Ejercicio 4

Se implementa el método de Euler en su forma vectorial.

```julia
function f(x,t,k) #Se crea una función que es un vector que servirá para implementar el método de Euler en su 
    #forma vectorial.
    return [x[2],-k*x[1]] #Se regresan las componentes del vector. 
end
```

```julia
function metodo_euler_vectorial(f,x0,t0,tf,h,k) #Se crea una función que permite, mediante el método vectorial
    #de Euler, obtener la solución aproximada de una ecuación diferencial. Dicha función toma como entradas la 
    #función que aparece en tal ecuación (la que creamos anteriormente), la condición inicial, el valor inicial
    #de t, el valor final de t, el valor de h y el valor de k.
    listt = [] #Se crea un vector cuyo valor inicial es t0 y valor final tf donde la separación entre  
    #sus valores está dada por el valor de h.
        for i in t0:h:tf
            push!(listt,i)
        end
    listx = [] #Se crea un vector en donde se guardan los valores xk del método de Euler.
    xk = x0 
        for j in 1:length(listt)
            push!(listx,xk)
        xk = xk + f(xk,listt[j],k)*h         
        end
    return listt, listx #Se regresan la listas listt y listx.
end
```

```julia
metodo_euler_vectorial(f,[0,1],0,8,0.5,2) #Se realiza un ejemplo particular del método vectorial de Euler. 
```
